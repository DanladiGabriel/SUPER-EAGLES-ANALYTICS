```python
SE.isna().sum()
```

minutes per match for each player

```python
SE['MINS_PER_MATCH'] = SE['MINUTES PLAYED'] / SE['MATCHES 2021/2022'].astype(int)
```

#Goals scored per match

```python
SE['GOALS_PER_MATCH'] = SE['GOAL SCORED'] / SE['MATCHES 2021/2022'].astype(float)
```

```python
SE.head()
```

TOTAL GOALS SCORED BY THE SUPER EAGLES 2021/2022

```python
total_goals = SE['GOAL SCORED'].sum()
print(total_goals)
```

PENALTY GOALS SCORED

```python
penalty_goals = SE['PENALTY SCORED'].sum()
penalty_goals
```

PENALTY ATTEMPTED

```python
total_penalty_attempts = SE['PENALTY ATTEMPTED'].sum()
total_penalty_attempts
```

MISSED PENALTY

```python
penalty_missed = total_penalty_attempts - penalty_goals
penalty_missed
```

PIECHART FOR PENALTY MISSED VS PENALTY SCORED

```python
plt.figure(figsize = (13, 6))
pl_not_scored = penalty_missed
data = [pl_not_scored, penalty_goals]
labels = ["penalty_missed", "penalty_goals"]
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
PLAYERS BASED ON AGE GROUP
```

```python
under20 = SE[SE['AGE'] <= 20]
age20_25 = SE[(SE['AGE'] >20) & (SE['AGE'] <= 25)]
age25_30 = SE[(SE['AGE'] >25) & (SE['AGE'] <= 30)]
above30 = SE[SE['AGE'] > 30]
```

```python
x = np.array([under20['NAMES OF PLAYERS'].count(), age20_25['NAMES OF PLAYERS'].count(), age25_30['NAMES OF PLAYERS'].count(), above30['NAMES OF PLAYERS'].count()])
mylabels = ['under 20', '20-25', '25-30', 'above 30']
plt.title('players based on age group', fontsize = 20)
plt.pie(x, labels = mylabels, autopct = '%.1f%%')
plt.show()
```

UNDER 25 PLAYERS IN EACH CLUB

```python
players_under_25 = SE[SE['AGE'] < 25]
players_under_25['CLUB'].value_counts().plot(kind = 'bar', color = sns.color_palette('cubehelix'))
```

PLAYERS UNDER 25 IN WATFORD

```python
players_under_25[players_under_25['CLUB'] == "WATFORD FC"]
```

PLOT OF AVERAGE AGE OF PLAYERS IN EACH CLUB

```python
plt.figure(figsize = (12, 6))
sns.boxplot(x = 'CLUB', y = 'AGE', data = SE)
plt.xticks(rotation = 90)
```

AVERAGE AGE OF SUPER EAGLES PLAYERS BY CLUB

```python
num_player = SE.groupby('CLUB').size()
data = (SE.groupby('CLUB')['AGE'].sum()) / num_player
data.sort_values(ascending = False)
```

AVERAGE AGE OF THE SUPER EAGLES SQUAD

```python
average_age = SE['AGE'].sum() / np.size(SE['S/N'])
average_age
```

total assist from each club visual

```python
assists_by_clubs = pd.DataFrame(SE.groupby("CLUB", as_index = False)['ASSISTS'].sum())
sns.set_theme(style = "whitegrid", color_codes = True)
ax = sns.barplot(x = 'CLUB', y = 'ASSISTS', data = assists_by_clubs.sort_values(by = 'ASSISTS'))
ax.set_xlabel('CLUB', fontsize = 30)
ax.set_ylabel('ASSISTS', fontsize = 20)
plt.xticks(rotation = 75)
plt.rcParams['figure.figsize'] = (15,8)
plt.title('plot of clubs vs total Assists', fontsize = 20)
```
